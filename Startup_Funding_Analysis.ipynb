**Data processing**

- Clean column names

```python
df.columns=df.columns.str.strip()
```

- Convert Entry column to datetime

```python
df['Entry_Date']=pd.to_datetime(df['Entry'],format='%b/%Y')
df['Entry_Year']=df['Entry_Date'].dt.year
df.head()
```

```python
df.dtypes
```

- Create valuation growth metric

```python
df['Valuation_Growth']=df['Valuation ($B)']/df['Entry Valuation^^ ($B)']
df.head()
```

- Extract primary city(before '/')

```python
df['Primary_City']=df['Location'].str.split('/').str[0]
df.head()
```

- Count numbers of Investors

```python
df['Investor_Count']=df['Select Investors'].str.count(',')+1
df.head()
```

## **Univariate Analysis**

- Valuation Distribution Analysis

```python
plt.hist(df['Valuation ($B)'],bins=20,alpha=0.7,color='skyblue',edgecolor='red')
plt.title('Distribution of Current Valuation')
plt.xlabel('Valuation ($B)')
plt.ylabel('Frequency')
plt.show()
```

- Sector Distribution

```python
plt.figure(figsize=(12,8))

sector_counts=df['Sector'].value_counts()
colors=plt.cm.Set3(np.linspace(0,1,len(sector_counts)))

plt.pie(sector_counts.values,labels=sector_counts.index,colors=colors,autopct='%1.1f%%',startangle=35)
plt.title('Distribution of Unicorns of Sectors',fontsize=16,fontweight='bold')
plt.axis('equal')
plt.show()
```

```python
def display_top_k_sectors(df,k):
    sector_counts=df['Sector'].value_counts()

    #Select top K sectors and group the rest as 'othes'
    if k<len(sector_counts):
      top_sectors=sector_counts.head(k)
      othes_sectors_count=sector_counts.iloc[k:].sum()
      top_sectors['Other']=othes_sectors_count
    else:
      top_sectors=sector_counts

    plt.figure(figsize=(12,8))
    colors=plt.cm.Set3(np.linspace(0,1,len(top_sectors)))

    plt.pie(top_sectors.values,labels=top_sectors.index,colors=colors,autopct='%1.1f%%',startangle=90)
    plt.title(f'Distribution of Top {k} Unicorns of Sectors',fontsize=16,fontweight='bold')
    plt.axis('equal')
    plt.show()
#Call the function with K=5 to display top 5 sectors
display_top_k_sectors(df,5)
```

- Top sectors bar sectors

```python
plt.figure(figsize=(12,6))
top_sectors=sector_counts.head(10)
bars=plt.bar(range(len(top_sectors)),top_sectors.values,color='skyblue')
plt.title('Top 10 Sectors by Number of Unicorns',fontsize=14,fontweight='bold')
plt.xlabel('Sector')
plt.ylabel('Number of Unicorns')
plt.xticks(range(len(top_sectors)),top_sectors.index,rotation=40,ha='right')

#Add value labels on bars
for bar,value in zip(bars,top_sectors.values):
  plt.text(bar.get_x()+bar.get_width()/2,bar.get_height()+0.1,str(value),ha='center',va='bottom')


plt.tight_layout()
plt.show()

```

**key insights:**

- EdTech leads with highest number of unicorns
- E-commerce and SaaS are major contributors
- Edtech emergence as significant sector
- Diversification across multiple sectors

# **Geographic Distribution**

- City-wise distribtion

```python
plt.figure(figsize=(14,6))

city_counts=df['Primary_City'].value_counts().head(10)
bars=plt.bar(city_counts.index,city_counts.values,color='pink')
plt.title('Top 10 Cities with Most Unicorns',fontsize=14,fontweight='bold')
plt.xlabel('Cities')
plt.ylabel('Number of Unicorns')
plt.xticks(rotation=90)

#Add value labels
for bar,value in zip(bars,city_counts.values):
  plt.text(bar.get_x()+bar.get_width()/2,bar.get_height()+0.1,str(value),ha='center',va='bottom')

plt.tight_layout()
plt.show()
```

- Creat a summary table

```python
city_summary=df.groupby('Primary_City').agg({
    'Valuation ($B)':['count','sum','mean'],
    'Entry_Year':['min','max']
}).round(2)
city_summary.columns=['Count','Total_Valuation','Avg_Valuation','First Entry','Last Entry']
city_summary.head(10)
```

**Key Insights:**
  - Banglore Dominates with highest number of unicorns
  - Mumbai and Gurgaon are other major hubs
  - Geographic concetration in tech cities
  - Banglore leads in both count and total valuation

# **Temporal Analysis**

   - Entry Year Trends

```python
yearly_entries=df['Entry_Year'].value_counts().sort_index()
bars=plt.bar(yearly_entries.index,yearly_entries.values,color='pink')
plt.title('Number of Unicorns Entered by year',fontsize=14,fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Number of Unicorns')
plt.xticks(yearly_entries.index,rotation=45)

#Add value labels
for bar,value in zip(bars,yearly_entries.values):
  plt.text(bar.get_x()+bar.get_width()/2,bar.get_height()+0.1,str(value),ha='center',va='bottom')

plt.tight_layout()
plt.show()
```

- Cumulative unicorns

```python
cumulative=yearly_entries.cumsum()
plt.plot(cumulative.index,cumulative.values,marker='o',linestyle='--',color='blue')
plt.title('Cumulative Unicorns over Time',fontweight='bold')
plt.xlabel('Years')
plt.ylabel('Cumulative Count')
plt.grid(True,alpha=0.3)
plt.show()
```

- Average valuation by year

```python
avg_val_by_year=df.groupby('Entry_Year')['Valuation ($B)'].mean()
plt.bar(avg_val_by_year.index, avg_val_by_year.values)
plt.xlabel('Entry Year')
plt.ylabel('Average Valuation ($B)')
plt.title('Average Valuation by Entry Year',fontweight='bold')
plt.show()
```

- sector evolution over time

```python
def display_top_k_sectors_evolution(df,k):
   #Group by year and sector,count occurrences
   sector_year=df.groupby(['Entry_Year','Sector']).size().unstack(fill_value=0)

   #Select top K sector based on overall count
   top_k_sectors_names=df['Sector'].value_counts().head(k).index.tolist()

   sector_year=sector_year[top_k_sectors_names]

   plt.figure(figsize=(15,8))
   sector_year.plot(kind='bar',alpha=0.7,ax=plt.gca())

   plt.title(f'Evulotion of Top {k} Sectors Over Time',fontsize=16,fontweight='bold')
   plt.xlabel('Entry Year')
   plt.ylabel('Number of Unicorns')
   plt.legend(title='Sector',bbox_to_anchor=(1.1,1.02),loc='upper left')
   plt.yticks(range(0,sector_year.max().max()+1,5))
   plt.grid(True,alpha=0.3)
   plt.tight_layout()
   plt.show()

#call the function with K=5 to display top 5 sector
display_top_k_sectors_evolution(df,5)
```
