# NumPy

- creating arrays

```python
import numpy as np
a=np.array([5,6,7,8,9])
b=np.arange(2,7)
c=np.zeros((2,3))
d=np.ones((3,2))


print("Array a:",a)
print("Array b:",b)
print(" 2x3 zeros:",c)
print("3x2 ones:",d)
```

- Indexing & slicing

```python
print("First element of a:",a[0])
print("Last tow element of a:",a[-2:])
```

**Basic arithmetic operation**

```python
print("a+ 2:",a+2)
print("a*3:",a*3)
```

Basic statistics

```python
print("Square of a:",np.square(a))
print("Mean of a:",np.mean(a))
print("Standard deviation:",np.std(a))
```

Reshaping Arrays

```python
matrix =np.arange(1,10).reshape((3,3))
print("Reshaped 3x3 matrix:")
matrix
```

Tranpose

```python
print("Transpose:")
matrix.T
```

Boolean indexing

```python
print("Elements >= 5:",matrix[matrix >= 5])
```

```python
import pandas as pd

data={
    'Name':['Alice','Bob','Charlie','David','Eva'],
    'Age':[25,30,35,40,22],
    'Department':['Sales','Engineering','HR','Engineering','Sales'],
    'Salary':[50000,80000,40000,90000,45000]
}
df=pd.DataFrame(data)
df
```

Basic operations

```python
print("First 3 rows:")
df.head(3)
```

```python
print("Column name:")
df.columns
```

```python
print("Data Types:")
df.dtypes
```

```python
print("describe numerical columns:")
df.describe()
```

Filtering

```python
print("People with salary >=50000:")
df[df['Salary']>=50000]
```

Sorting

```python
print("Sorted by Age:")
df.sort_values(by='Age')
```

Grouping and aggregation

```python
grouped=df.groupby('Department')['Salary'].mean()
print("Average salary by Department:")
grouped
```

Adding new columns

```python
df['Salary in Lakhs']=df['Salary']/100000
df
```

Line Plot

```python
import matplotlib.pyplot as plt


years = [2019,2020,2021,2022]
revenue = [100,110,130,160]

plt.plot(years,revenue,marker='o',linestyle='--',color='red')
plt.title("Company Revenue over Years")
plt.xlabel("Years")
plt.ylabel("Revenue (in crores)")
plt.xticks(years)
plt.grid(True)
plt.show()
```

Bar chart

```python
departments =df['Department'].value_counts()
departments.plot(kind='bar',color='teal')
plt.title("Employee Count by Department")
plt.xlabel("Department")
plt.ylabel("Count")
plt.show()
```

Histogram

```python
ages=[20,55,40,34,24,21,26]
plt.hist(ages,bins=5,edgecolor='black',color='coral')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

pie chart

```python
df.groupby('Department').size().plot(kind='pie',autopct='%1.1f%%')
plt.title("Department Proportion")
plt.ylabel("")
plt.show()
```

Seaborn

Load simple dataset

```python
import seaborn as sns

tips=sns.load_dataset("tips")
tips.head()
```

```python
print("\nInfo about the tips dataset")
tips.info()

print("\nDescription of the tips dataset")
tips.describe(include='all')
```

```python
sns.pairplot(tips,hue="sex")
plt.show()
```

```python
corr =tips.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.title("correlation matrix")
plt.show()
```

```python
sns.countplot(x='day',data=tips,palette='Set1',hue='day',legend=True)
plt.title("Count of days")
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title("Boxplot of total bill by day")
plt.show()
```
